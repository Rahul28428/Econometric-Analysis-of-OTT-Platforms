```python
import numpy as np
import matplotlib.pyplot as plt

# Generate linear function y = x
xi, yi = 7, 7  # End point on the line y = x
x_linear = np.linspace(0, xi,7)
y_linear = x_linear

# Generate noisy wave-like function starting from (0, 0) and ending at (xi, yi)
np.random.seed(42)  # for reproducibility
noise = np.random.normal(scale=1, size=len(x_linear))
y_wavey = np.cumsum(noise)  # Cumulative sum to create a wave-like pattern
y_wavey = y_wavey - y_wavey.min()  # Shift the pattern to start from zero

# Scale the wave to exactly reach the specified point (xi, yi)
y_wavey = y_wavey * (yi / y_wavey[-1])  # Scale to end at the specified point (xi, yi)

# Plotting
plt.plot(x_linear, y_linear, label='y = x', color='blue')
plt.plot(x_linear, y_wavey, label='Noisy Wavey', color='orange')

plt.scatter(xi, yi, color='red')  # Mark the end point on y = x

plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear and Noisy Wavey Functions')
plt.legend()
plt.grid(True)
plt.show()
```

##Linear Regession analysis

```python
import pandas as pd
import statsmodels.api as sm

# Provided data
data = {
    'Platform': ['Disney+', 'Netflix', 'SonyLiv', 'Amazon Prime Video', 'Zee5', 'Voot', 'TVF Play', 'MX Player', 'Eros Now', 'ALTBalaji'],
    'Subscribers': [61.3, 36.23, 32.7, 28, 48.11, 1, 11.3, 78, 36.2, 35],
    'Market_shares': [50, 8, 6, 9, 5, 5, 4, 2.6, 6, 4],
    'Revenue (in lakh Rs.)': [13875860, 13875860, 1413322, 5034.115, 549600, 1744869, 4755.414, 86378.927, 1578691, 36800]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Simple Linear Regression for Subscribers
X_subscribers = sm.add_constant(df['Subscribers'])
model_subscribers = sm.OLS(df['Revenue (in lakh Rs.)'], X_subscribers).fit()
print(model_subscribers.summary())

# Simple Linear Regression for Market Shares
X_market_shares = sm.add_constant(df['Market_shares'])
model_market_shares = sm.OLS(df['Revenue (in lakh Rs.)'], X_market_shares).fit()
print(model_market_shares.summary())

# Multiple Regression with both Subscribers and Market Shares
X_multiple = sm.add_constant(df[['Subscribers', 'Market_shares']])
model_multiple = sm.OLS(df['Revenue (in lakh Rs.)'], X_multiple).fit()
print(model_multiple.summary())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot for Simple Linear Regression with Subscribers
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
sns.scatterplot(x=df['Subscribers'], y=df['Revenue (in lakh Rs.)'])
plt.title('Subscribers vs Revenue')
plt.xlabel('Subscribers')
plt.ylabel('Revenue (in lakh Rs.)')
sns.regplot(x=df['Subscribers'], y=df['Revenue (in lakh Rs.)'], ci=None, line_kws={'color':'red'})

# Scatter plot for Simple Linear Regression with Market Shares
plt.subplot(1, 3, 2)
sns.scatterplot(x=df['Market_shares'], y=df['Revenue (in lakh Rs.)'])
plt.title('Market Shares vs Revenue')
plt.xlabel('Market Shares')
plt.ylabel('Revenue (in lakh Rs.)')
sns.regplot(x=df['Market_shares'], y=df['Revenue (in lakh Rs.)'], ci=None, line_kws={'color':'red'})

# Scatter plot for Multiple Regression
plt.subplot(1, 3, 3)
ax = sns.scatterplot(x=df['Subscribers'], y=df['Revenue (in lakh Rs.)'], hue=df['Market_shares'])
plt.title('Linear regression')
plt.xlabel('Subscribers')
plt.ylabel('Revenue (in lakh Rs.)')
sns.regplot(x=df['Subscribers'], y=df['Revenue (in lakh Rs.)'], scatter=False, ax=ax, ci=None, line_kws={'color':'red'})

plt.tight_layout()
plt.show()
```

##Lasso/ Ridge Regression


```python
import pandas as pd
from sklearn.linear_model import Lasso, Ridge

# Provided data
data = {
    'Platform': ['Disney+', 'Netflix', 'SonyLiv', 'Amazon Prime Video', 'Zee5', 'Voot', 'TVF Play', 'MX Player', 'Eros Now', 'ALTBalaji'],
    'Subscribers': [61.3, 36.23, 32.7, 28, 48.11, 1, 11.3, 78, 36.2, 35],
    'Market_shares': [50, 8, 6, 9, 5, 5, 4, 2.6, 6, 4],
    'Revenue (in lakh Rs.)': [13875860, 13875860, 1413322, 5034.115, 549600, 1744869, 4755.414, 86378.927, 1578691, 36800]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Multiple Regression with both Subscribers and Market Shares (OLS)
X_multiple = pd.get_dummies(df[['Subscribers', 'Market_shares']], drop_first=True)
X_multiple = sm.add_constant(X_multiple)
model_multiple_ols = sm.OLS(df['Revenue (in lakh Rs.)'], X_multiple).fit()

# Lasso Regression
lasso_model = Lasso(alpha=1.0)
lasso_model.fit(X_multiple, df['Revenue (in lakh Rs.)'])

# Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_multiple, df['Revenue (in lakh Rs.)'])

# Display coefficients
results_df = pd.DataFrame({
    'Variable': model_multiple_ols.params.index,
    'OLS Coefficients': model_multiple_ols.params.values,
    'Lasso Coefficients': lasso_model.coef_,
    'Ridge Coefficients': ridge_model.coef_,
})

print(results_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import StandardScaler

# Data
X = df[['Subscribers', 'Market_shares']]
y = df['Revenue (in lakh Rs.)']

# Standardize the features
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)
linear_model = LinearRegression()
linear_model.fit(X_standardized, y)
linear_predictions = linear_model.predict(X_standardized)
# Lasso Regression
lasso_model = Lasso(alpha=1.0)
lasso_model.fit(X_standardized, y)
lasso_predictions = lasso_model.predict(X_standardized)

# Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_standardized, y)
ridge_predictions = ridge_model.predict(X_standardized)

# Scatter plot for Lasso Regression
plt.figure(figsize=(16, 4))
# Scatter plot for Linear Regression
plt.subplot(1, 4, 1)
sns.scatterplot(x=linear_predictions, y=y)
plt.title('Linear Regression')
plt.xlabel('Predicted Revenue (Linear)')
plt.ylabel('Actual Revenue')
sns.regplot(x=linear_predictions, y=y, ci=None, line_kws={'color':'red'})
plt.subplot(1, 4, 2)
sns.scatterplot(x=lasso_predictions, y=y)
plt.title('Lasso Regression')
plt.xlabel('Predicted Revenue (Lasso)')
plt.ylabel('Actual Revenue')
sns.regplot(x=lasso_predictions, y=y, ci=None, line_kws={'color':'blue'})

# Scatter plot for Ridge Regression
plt.subplot(1, 4, 3)
sns.scatterplot(x=ridge_predictions, y=y)
plt.title('Ridge Regression')
plt.xlabel('Predicted Revenue (Ridge)')
plt.ylabel('Actual Revenue')
sns.regplot(x=ridge_predictions, y=y, ci=None, line_kws={'color':'green'})

# Scatter plot for Multiple Regression
plt.subplot(1, 4, 4)
ax = sns.scatterplot(x=lasso_predictions, y=y, hue=ridge_predictions)
plt.title('Lasso and Ridge Regression')
plt.xlabel('Predicted Revenue ')
plt.ylabel('Actual Revenue')
sns.regplot(x=lasso_predictions, y=y, scatter=False, ax=ax, ci=None, line_kws={'color':'red'},label='Lasso Regression')
sns.regplot(x=ridge_predictions, y=y, scatter=False, ax=ax, ci=None, line_kws={'color':'blue'}, label='Ridge Regression')

plt.tight_layout()
plt.show()
```

```python
# Horizontal Bar plot for Subscribers
plt.figure(figsize=(10, 6))
data['Subscribers(in millions)'].sort_values().plot(kind='barh', color='skyblue')
plt.title('Subscribers by Platform')
plt.xlabel('Subscribers (in millions)')
plt.ylabel('Platform')
plt.tight_layout()
plt.show()

# Pie chart for Market Share
plt.figure(figsize=(8, 8))
plt.pie(data['Market_shares(in % age)'], labels=data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Market Share Distribution')
plt.show()

# Donut chart for Revenue
plt.figure(figsize=(8, 8))
plt.pie(data['Revenue (in Rs.)'], labels=data.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'), wedgeprops=dict(width=0.3))
plt.title('Revenue Distribution')
plt.show()
```
